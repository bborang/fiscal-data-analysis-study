## 재정데이터를 활용한 분석 실습(1)

- 목적 : 지방자치단체별 일일 분야별 세출현황을 통한 예산액에 대한 시사점 도출

- 방법 : 예산현액과 지출액의 차이를 계산해 파생변수 '차액' 생성 / 지방자치단체 중 '서울'만 추출 / '서울' 지역에 따른 '분야명'을 기준으로 차액 비교 시각화

- 분석 데이터셋 : 분야별 세출현황(3년간, 2021~2023년) 데이터셋 (출처 : 열린재정, 지방재정 365)

```python
## 2021년 세출현황 데이터셋 전처리

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 과학적 표기법 해제
pd.options.display.float_format = '{:.0f}'.format

df1 = pd.read_csv("Practice Data/분야별 세출현황_2021.csv")
df1.info() #결측치 없음

df1 = df1[['회계연도','지역명','분야명','예산현액','지출액']]
df1.head()

# 특정 컬럼의 항목('서울') 추출
df1[df1['지역명'] == '서울' ]
df1['지역명'].unique()
df1.head()

# 파생변수 만들기
df1['차액'] = df1['예산현액'] - df1['지출액']
df1.head()

# 필요한 컬럼만 추출(2)
df1 = df1[['회계연도','분야명','차액']]
df1.head()
```

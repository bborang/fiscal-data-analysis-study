## matplotlib 
- 파이썬에서 데이터를 시각화해주는 라이브러리

### 막대그래프
- 막대그래프(Bar graph, Bar chart)는 범주가 있는 데이터 값을 직사각형의 막대로 표현하는 그래프
- matplotlib에서는 matplotlib.pyplot 모듈의 bar() 함수 사용

### 히스토그램
- 히스토그램은 도수분포표를 그래프로 나타낸 것으로, 가로축 : 계급 / 세로축 : 도수 를 표현
- matplotlib.pyplot 모듈의 hist() 함수 이용
- 구간의 크기를 지정, bins = 구간 갯수

```python
# 세로 막대 그래프
import matplotlib.pyplot as plt

day = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
work_time = [2,6,7,5,7,8,3]

plt.figure(figsize=(5,3)) # 그래프의 크기

plt.bar(day, work_time, color='green', edgecolor='blue', alpha=0.4) 
plt.xticks(rotation=45) # x축 눈금 라벨 회전(글자수에 따라 선택)
plt.xlabel('Day') # x축 라벨
plt.ylabel('Time') # y축 라벨  
plt.title('Working Time') # 그래프 제목
plt.ylim(0,10) # y축 범위
plt.show() # 그래프 출력

```

```python
# 가로 막대 그래프
import matplotlib.pyplot as plt

day = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
work_time = [2,6,7,5,7,8,3]

plt.figure(figsize=(5,3)) # 그래프의 크기

plt.barh(day, work_time, color='green', edgecolor='blue', alpha=0.4) 
plt.yticks(rotation=45) # x축 눈금 라벨 회전(글자수에 따라 선택)
plt.ylabel('Day') # x축 라벨
plt.xlabel('Time') # y축 라벨  
plt.title('Working Time') # 그래프 제목
plt.xlim(0,10) # x축 범위
plt.show() # 그래프 출력

```

```python
# 히스토그램
import matplotlib.pyplot as plt

score = [50,60,70,80,90,55,65,75,85,95,52,62,72,82,92,11,12,13,14,15]

plt.figure(figsize=(8, 5))

plt.hist(score, bins=10, label='Score Distribution', color='skyblue', edgecolor='white')

plt.title('Score Histogram')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

```python
# 히스토그램(누적) 
# camulative = True : 누적그래프 | False : 기본값(누적x)
import matplotlib.pyplot as plt

score = [50,60,70,80,90,55,65,75,85,95,52,62,72,82,92,11,12,13,14,15]

plt.figure(figsize=(8, 5))

plt.hist(score, bins=10, color='skyblue', edgecolor='white', cumulative=True, label='cumulative = True')

plt.title('Score Histogram')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

```python
# 선 그래프(1)
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))

plt.plot([1,2,3,4],[2,3,4,5], 'g-')
plt.plot([1,2,3,4],[2,4,6,8],'r--')
plt.plot([1,2,3,4],[2,4,6,9], 'y:')
plt.plot([1,2,3,4],[1,2,3,4],'b-')

plt.show()
```

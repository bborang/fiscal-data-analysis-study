## 판다스(pandas)

- 데이터 조작과 분석을 위한 파이썬 소프트웨어 라이브러리
- csv, xls, json 등에 담긴 데이터를 데이터프레임이란 데이터타입으로 불러와서 작업 가능
- 데이터프레임은 2차원 데이터로 구성.

- 1차원 데이터 구조 : Series
- 2차원 데이터 구조 : DataFrame


- 행과 열의 변경 함수는 `df.transpose()` or `df.T`

- 저장 : `result = pd.to_파일형식('파일이름')`
- 데이터 불러오기 : `data = pd.read_파일형식('파일이름')`

```python
import pandas as pd

data = {'사원번호':range(1000,1010),  '평가점수':[10,20,10,14,15,16,55,65,60,70]}
df = pd.DataFrame(data)
print(df.head())
print(df.tail())
print(df.info())
print(df.shape)

#행과 열 변경
print(df)
print(df.T)

#인덱스 값 변경
index_set = ['A','B','C','D','E','F','G','H','I','J']
df = pd.DataFrame(data,index=index_set)
print(df)

#저장하기
result = df.to_csv("study.csv")
```

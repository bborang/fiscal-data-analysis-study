```python
# 세로 막대 그래프
import matplotlib.pyplot as plt
import seaborn as sns

day = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
work_time = [2,6,7,5,7,8,3]

plt.figure(figsize=(5,3)) # 그래프의 크기

sns.barplot(x= day, y = work_time, palette = 'pastel',hue=day) 
plt.xticks(rotation=45) # x축 눈금 라벨 회전(글자수에 따라 선택)
plt.xlabel('Day') # x축 라벨
plt.ylabel('Time') # y축 라벨  
plt.title('Working Time') # 그래프 제목
plt.ylim(0,10) # y축 범위
plt.show() # 그래프 출력

```

```python
# 가로 막대 그래프
# barplot() 함수의 orient='h' 옵션을 사용
import matplotlib.pyplot as plt
import seaborn as sns

day = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
work_time = [2,6,7,5,7,8,3]

plt.figure(figsize=(5,3)) # 그래프의 크기

sns.barplot(y= day, x = work_time, palette = 'bright',hue=day, orient='h') 
plt.xticks(rotation=45) # x축 눈금 라벨 회전(글자수에 따라 선택)
plt.xlabel('Day') # x축 라벨
plt.ylabel('Time') # y축 라벨  
plt.title('Working Time') # 그래프 제목

plt.show() # 그래프 출력

```

```python
#히스토그램 
# KDE(커널밀도추정) : 부드러운 곡선을 줌

import matplotlib.pyplot as plt
import seaborn as sns

score = [50,60,70,80,90,55,65,75,85,95,52,62,72,82,92,11,12,13,14,15]

plt.figure(figsize=(8, 5))

sns.histplot(data = score, bins=10, kde = True)

plt.title('Score Histogram')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

```python
# 산점도 

import matplotlib.pyplot as plt
import seaborn as sns

day = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
work_time = [2,6,7,5,7,8,3]

sns.scatterplot(x=day,y=work_time,palette='deep',hue=day,s=100)

plt.xlabel('Day') # x축 라벨
plt.ylabel('Time') # y축 라벨  
plt.title('Working Time') # 그래프 제목

plt.show() # 그래프 출력
```
